````python
import re
import os
import requests
from PIL import Image
from io import BytesIO

def display_image(response_str, display_size=200, max_size_kb=50):

    url = re.search(r'https://[^\s\)]+', response_str).group()
    img_data = requests.get(url).content
    img = Image.open(BytesIO(img_data))
  
    quality = 85
    while quality > 10:
        buffer = BytesIO()
        img.convert("RGB").save(buffer, format="JPEG", quality=quality)
        if buffer.tell() <= max_size_kb * 1024:
            break
        quality -= 5
    
    from IPython.display import Image as IPImage, display
    display(IPImage(data=buffer.getvalue(), width=display_size, height=display_size))


def save_image(response_str, filename):
    url = re.search(r'https://[^\s\)]+', response_str).group()
    img_data = requests.get(url).content
    img = Image.open(BytesIO(img_data))
    
    os.makedirs("img", exist_ok=True)
    
    # Save as JPG with compression to target ~100KB
    quality = 85
    while quality > 10:
        buffer = BytesIO()
        img.convert("RGB").save(buffer, format="JPEG", quality=quality)
        if buffer.tell() <= 100 * 1024:
            break
        quality -= 5
    
    with open("img/" + filename + ".jpg", "wb") as f:
        f.write(buffer.getvalue())
````
